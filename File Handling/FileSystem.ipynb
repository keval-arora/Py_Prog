```python
"""PATHLIB & PATH"""
from pathlib import Path
"""Directory & Path"""
################-DIRECTORY INFORMATIONS-#################
#Get Current working Directory:
cwd = Path.cwd()
print(type(cwd))
print("Current working dir:" , cwd)

#Get Parent Directories:
parent = cwd.parent
print("Parent working dir: ", parent)

#Get Child Directories in Current Directory:
for child in cwd.iterdir():
    if child.is_dir():
        print("Child in Current dir:  ", child)

for child in cwd.iterdir():   
    if child.is_file(): 
        print("Files in current dir:  ", child)
               
#Check if Dir or File:
print(parent.is_dir())
print(parent.is_file())

################-FILE INFORMATIONS-#################
#Create Full File Path Name:
new_file = Path(Path.joinpath(cwd, 'wr_file.txt'))
print("Full Path: ", new_file)

#Check if file exists:
print("Do File Exists: ", new_file.exists())

#Get File Name:
print("File Name: ", new_file.name)

#Get File extension:
print("File extension: ", new_file.suffix)

#Get Folder Name:
print("Folder Name: ", new_file.parent.name)

#Get File Size in Kbytes:
print("File Size: ", new_file.stat().st_size)

########## File Handling #########
stream = open('wr_file.txt', 'rt+')
print(type(stream))


#Check if file is readable
print(stream.readable())

#Read n Char
print(stream.read(1))

#Read entire line
print(stream.readline())

#Write File
stream.write('K')

#Write entire line which can be object or text
stream.writelines('EVAL, ARORA')

#Write Single character
stream.write('\n')

#Write entire line which can be object or text
stream.writelines('\n'.join(['Works in Optum']))

#Reset cursor to start 
stream.seek(0)

#Flushing
stream.flush()

#Close file
stream.close()

#try and finally alternative is with
with open('wr_file.txt', 'rt+') as stream:
    stream.write('\nHow are you')
    for record in stream:
        char1 = record.split()
        print(f"{char1}")
        


#With generally used to aquire resource, use that resource and auto close it when done.
#Resource can be file, database, network connection etc etc
#Files in Python are Text Files which is sequence of char or Binary Files which are sequence of bytes
#Standard Obj file in Ptyhon:
## sys.stdin- Standard Input Stream which reads input from command line or can be directed to read from text or other resource.
## sys.stdout- Standard Output Stream which writes output to command line or can be directed to write to text or other resource.
## sys.stderr- Standard Error Stream for Error Handling.
```
