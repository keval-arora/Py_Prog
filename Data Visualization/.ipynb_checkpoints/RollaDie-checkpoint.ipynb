```python
import random
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
roll_die = [random.randrange(1,7) for _ in range(int(input("Enter the number of times you want to roll a die")))]
val, freq = np.unique(roll_die, return_counts = True)
print(val, freq)

"""Creating Axes object and labeling the title, x-axis, y-axis"""
title = f"Rolling a die {len(roll_die):,} times"
sns.set_style('whitegrid')
axes = sns.barplot(x= val, y= freq, palette ='bright')
axes.set_title(title)
axes.set(xlabel= "Die Face", ylabel= "Frequency")
axes.set_ylim(top=max(freq)*1.20)
#axes.set_xlim(top=max(freq)*1.20)
"""Labeling each bar"""
for bar, val in zip(axes.patches, freq):
    print(bar,val)
    x_bar = bar.get_x()+bar.get_width()/2.0
    y_bar = bar.get_height()
    text = f' {y_bar}\n{y_bar/len(roll_die):.3%} '
    axes.text(x_bar, y_bar, text, fontsize = 11, ha='center',va='bottom')
```
