```python
"""Binary Trees"""

class Node():
    def __init__(self,key):
        self.data = key
        self.left_point = None
        self.right_point = None
        
        
class BST():
    def __init__(self):
        self.root = None
        
        
    def insert(self,key):
        if not isinstance(key, Node):
            key = Node(key)            
        if self.root == None:
            self.root = key
        else:
            self._insert(curr_root = self.root, key = key)
                
            
    def _insert(self,curr_root,key):
        """Inserting current value of key, root recursively"""
        if key.data < curr_root.data:
            if curr_root.left_point == None:
                curr_root.left_point = key
            else:
                self._insert(curr_root.left_point,key)          
        elif key.data > curr_root.data:
            if curr_root.right_point == None:
                curr_root.right_point = key
            else:
                self._insert(curr_root.right_point,key)
                         
    def in_order(self):
        self._in_order(self.root)
        
    def _in_order(self,curr_root):
        if curr_root:
            self._in_order(curr_root.left_point)
            print(curr_root.data)
            self._in_order(curr_root.right_point)
            
#     def pre_order(self):
#         pass
#     def _pre_order(self, curr):
#         pass
#     def post_order(self):
#         pass
#     def _post_order(self, curr):
#         pass

tree = BST()

tree.insert("F")                        #Root
print(tree.root.data)
tree.insert("C")                        #F's Left Child
print(tree.root.left_point.data)
tree.insert("G")                        #C's Left Child
print(tree.root.right_point.data)
tree.insert("A")                        #G's Left Child
print(tree.root.left_point.left_point.data)
tree.insert("B")                        #A's Right Child
print(tree.root.left_point.left_point.right_point.data)
tree.insert("K")
print(tree.root.right_point.right_point.data)
print("Now Traversing")
print(tree.in_order())

#Non-Linear/absbtract data structures:
#-A data with a pointer together is called nodes
#-Tree has root node with two/more pointers each pointing to child nodes
#-Bottom most nodes with no child is known as leaf Node
#-Tree Rules:
##-Trees dont have cycle like a-b-c-d-a is not possible
##-Node cannot have multiple parent like b and c cannot point to d at same time
##-Tree cannot have multiple root
##-Root in tree cannot point to itself or parent cannot be its own child
#-Number of levels taken to reach from root to leaf is known as height of tree
#-BINARY TREE can have only 2 nodes
#-In BINARY SEARCH TREE, child nodes which has greater value than parent are alligned right to its parent and the ones which has lesser value,
#### are alligned to the left
#Search, Insert, Delete are done with Time complexity of B(h) where h is height of a tree.
#Left/Right Skwed BST branches are formed while alligning greater / lesser values to right or left which act like linear singly linked list.
##So it has O(h) and not O(logn)
#In-order traversal of BST gives sorted representation of node from left, root, right

```
