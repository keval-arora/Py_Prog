```python
class HashMap():
    def __init__(self,size):
        self.table_size = size
        self.hash_table = [[] for _ in range(self.table_size)]
        
            
    def create_bucket(self, key):
        hashed_index = hash(key)%self.table_size
        hash_map_bucket = self.hash_table[hashed_index]
        return hash_map_bucket
    
    
    def set_val(self,key,value):
        found_key = False
        for ind, rec in enumerate(self.create_bucket(key)):
            rec_key, rec_val = rec
            if rec_key == key:
                found_key = True
                break            
        if found_key:
            self.create_bucket(key).insert(ind,(key,value))
        else:
            self.create_bucket(key).append((key,value))
                
    def get_val(self,key):
        found_key = False
        for ind, rec in enumerate(self.create_bucket(key)):
            rec_key, rec_val = rec
            print(rec)
            if rec_key == key:
                found_key = True
                break            
        if found_key:
            return rec_val
        else:
            return f"Value Not Found"
        
    def __str__(self):
        return " ".join(str(item) for item in self.hash_table)
    
hash_tab = HashMap(5)
print(hash_tab)
hash_tab.set_val('USA', 'us')
hash_tab.set_val('AUS', 'au')
hash_tab.set_val('Africa','af')
hash_tab.set_val('USA', 'us is great')
print(hash_tab)   
hash_tab.get_val('USA')
#print(hash_tab.get_val('USA'))    
```

```python
#__eq__() and __hash__()

#Making user defined class hashable: For desired behaviour uncomment lines below

class Symbol:
    def __init__(self, x):
        self.value = x
    
    def __eq__(self,other):
        if isinstance(self, other.__class__):
            return self.value == other.value
        else:
            return NotImplemented
        
    def __hash__(self):
        return hash(self.value)
               
    def __str__(self):
        return f"Value is {self.value}"   
    
s1 = Symbol(1)
s2 = Symbol(1)
print(s1 is s2)
print(s1 == s2)

set1 = set()

set1.add(s1)
set1.add(s2)

print(len(set1))
    
```
